```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
```

```python
t = np.linspace(0,np.pi*100,np.int(1e5))
N = 10+np.sin(t)
bkgd = stats.norm.rvs(size=np.int(1e5))*np.sqrt(N)+N
```

### Problem 2
Density now you can make a scatter plot of just a 100,000 points, so you might try it to see what it looks like. But particularly in the central areas the points all overlap into just a blob, and this just gets worse as the number of points goes up. What is better is to plot the density of the points. There are several ways of doing this, but a 2D histogram plot is the easiest (plt.hist2d() in python and histogram2() in MatLab).

2a) Make a 2d histogram plot. Read about options, and play with binsize so it is useful (too large a bin and everything washes out, too small and the sample noise dominates). Make sure you use a colorbar.

2b) Clearly explain what is being plotted in your plot (e.g. what does each color mean?)

```python
#Problem 2, Part A
plt.hist2d(t,bkgd,bins=(100,100))
plt.colorbar()
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('2D Histogram of Full Data, 100,000 Points')
plt.show()
```

Answer Question 2, Part B:
This 2D histogram plots all 100,000 data points, and the color represents how frequently a particular point is repeated, in other words how common the data point is. Points that fall in the purple/dark blue region are very rare, ranging from zero to about fifteen instances of that point. More common points appear in the light green/yellow region which is the center band region of the plot. These points occur between 45 and 60 times out of 100,000 in the data. Since the y values were randomly generated as a normal distribution, it makes sense that points in the middle of the specified range are more common, and thus are more common than y values that are on the more extreme ends of the specified range.

### Problem 3
Folding now if you have a periodic signal, it is possible to fold the data stream at that period to make the pattern clearer.

3a) Make an new time variable that folds at a period of 2π (the modulo operator is your friend, variably expressed as mod(), %, or rem()). Then make a scatter plot of all your data, but now folded. (Reminder, for more data you'd never actually want to plot all the points.)

3b) Make a 2D histogram plot of your folded data.

```python
#Problem 3, Part A
tnew = t%(2*np.pi)
plt.plot(tnew,bkgd)
plt.xlabel('Time, Folded Period of 2 Pi')
plt.ylabel('Value')
plt.title('Folded Data, Value vs. Time')
plt.show()
```

```python
#Problem 3, Part 2
plt.hist2d(tnew,bkgd,bins=100)
plt.xlabel('Time, Folded Period of 2 Pi')
plt.ylabel('Value')
plt.title('2D Histogram of Value vs. Time, Folded Data at Period of 2 Pi')
plt.colorbar()
plt.show()
```
