```python
xlow=0
xhigh=14
xspace=15
x=np.linspace(xlow,xhigh,xspace)

pois_dist=scipy.stats.poisson.pmf(x,mu=4,loc=0)
pois_dist=pois_dist/sum(pois_dist)
comb=pois_dist


k=5
for i in range(k):
    comb=np.convolve(comb,pois_dist)
    
plt.step(x,pois_dist)
plt.yscale('Log')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Probability mas Function for Poisson Distribution: Day 1')
plt.show()


cx=np.linspace(xlow*k,xhigh*k,len(comb))
plt.step(cx,comb)
plt.yscale('Log')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Probability mas Function for Poisson Distribution: Day 5')
plt.show()
```
